```python
import pandas as pd
import numpy as np
df = pd.read_csv("weather_by_cities.csv",usecols=['day','city','temperature','windspeed'])
df
```

**Pivot function allows to reorganise the data according to required parameters**

```python
df.pivot(index='city',columns='day')
```

```python
df.pivot(index='city',columns='day',values='windspeed')
```

```python
df.pivot(index='day',columns='city')
```

# Pivot table

```python
df.pivot_table(index='day',columns='city')
```

```python
df.pivot_table(index='city',columns='day',margins=True,aggfunc=np.sum)
```

```python
df['day']=pd.to_datetime(df['day'])
```

```python
type(df['day'][0])
```

```python
df.pivot_table(index=pd.Grouper(freq='M',key='day'),columns='city')
```

```python
df.pivot_table(index=pd.Grouper(freq='M',key='day'),columns='city',values='temperature')
```
