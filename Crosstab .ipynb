```python
import pandas as pd
import numpy as np
df= pd.read_csv('weather_by_cities.csv')
df
```

```python
pd.crosstab(df.city,df.event) 
```

```python
pd.crosstab(df.city,df.event,margins= True)
```

### Multiindex columns and rows

```python
pd.crosstab([df.city],[df.temperature,df.event], margins= True)
```

```python
pd.crosstab(df.city,df.event,normalize='index') #gives the percentage
```

```python
pd.crosstab(df.event,df.city, values=df.temperature, aggfunc=np.average)
```
